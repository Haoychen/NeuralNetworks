```python
sorted(dicton.keys())
```

```python
print(sc.spatial.distance.cosine(dicton['airline'] - dicton['airlines'], dicton['airport'] - dicton['airports']) )
print(sc.spatial.distance.cosine(dicton['rent'] - dicton['rental'], dicton['serve'] - dicton['service']) )
print(sc.spatial.distance.cosine(dicton['book'] - dicton['booking'], dicton['connect'] - dicton['connecting']) )
print(sc.spatial.distance.cosine(dicton['canada'] - dicton['canadian'], dicton['america'] - dicton['american']) )
print(sc.spatial.distance.cosine(dicton['arizona'] - dicton['phoenix'], dicton['colorado'] - dicton['denver']) )
print(sc.spatial.distance.cosine(dicton['earlier'] - dicton['earliest'], dicton['later'] - dicton['latest']) )
print(sc.spatial.distance.cosine(dicton['arrive'] - dicton['canada'], dicton['arrive'] - dicton['america']) )
print(sc.spatial.distance.cosine(dicton['san'] - dicton['diego'], dicton['san'] - dicton['francisco']) )
print(sc.spatial.distance.cosine(dicton['serving'] - dicton['dallas'], dicton['serving'] - dicton['denver']) )
print(sc.spatial.distance.cosine(dicton['night'] - dicton['dinner'], dicton['morning'] - dicton['breakfast']) )
print(sc.spatial.distance.cosine(dicton['from'] - dicton['kansas'], dicton['from'] - dicton['houston']) )


```

I used cosine distance to evaluate the similarity. Based on the above results, I find that the similarity is more based on the context: such as F(night) - F(dinner) = F(morning) - F(breakfast); F(san) - F(diego) = F(san) - F(francisco). While the word paris, such as rent v.s. rental is not similar to serve v.s. service.

The interesting word pairs includes: from kansas - from houston; night dinner - morning breakfast; serving dallas - serving denver; san diego - san francisco; arrive canada - arrive america
